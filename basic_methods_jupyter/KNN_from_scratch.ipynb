```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class KNN:
    def __init__(self, k):
        self.k = k
    def fit(self, x, y):
        self.x = x
        self.y = y
    def predict(self, data):
        """
        data is N x F
        """
        lbl = np.zeros((data.shape[0],))
        for i, d in enumerate(data):
            dist = np.sum((d - self.x) ** 2, axis=-1)
            ind_knn = np.argsort(dist)[:k]
            lbl_knn = self.y[ind_knn]
            lbl[i] = np.mean(lbl_knn) > 0.5 + 0
        return lbl
        
```

```python
N = 50
x = np.random.randn(N, 2)
y = np.random.randint(2, size=N)
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
```

```python
data = np.array([[-1, -1.5], [1, -0.5], [-1, 0]])
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.scatter(data[:, 0], data[:, 1])
```

```python
knn = KNN(3)
knn.fit(x, y)
lbl_data = knn.predict(data)
print(lbl_data)
```
